Estacionalidad 


```python
import pandas as pd 
import numpy as np
%matplotlib inline 
```

```python
airline = pd.read_csv('airline_passengers.csv', index_col=0, parse_dates= True)
```

```python
airline
```

```python
airline.plot() #Se observa: estacionalidad en el conjunto de datos(picos y bajadas conforman la tendencia en alza), y una tendecia posiblemente exponencial, segÃºn el mes del aÃ±o hay fluctuaciones que se repiten, en los meses de vacaciones hay picos
```

Para descomponer los efectos:
1. Aditivo si la tendencia parece ser mÃ¡s lineal.
2. Multiplicativo si la tendencia parece ser exponencial.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(airline['Thousands of Passengers'], model='additive')
result.plot();

#trend: patrÃ³n de crecimiento general
#estacionalidad(seasonal): se repite aÃ±o tras aÃ±o
#residual= es lo que no se explica ni por tendencia o estacionalidad, parece ruido, 
#y es un indicador de donde hay mÃ¡s ruido en el cojunto de datos
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'], model='multiplicative')
result.plot();
```
